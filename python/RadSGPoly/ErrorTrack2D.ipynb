### Jacobian Resolution Test

compute derivates of errors with top luminosity for different step sizes.  Problems at both big and small step sizes arise. Big is normal, small must be an issue with the shooting.

```python
lL, lDt = guessHist[-1]
err0 = s2.error2D((lL, lDt), *eargs)
nD = 29 #chosen to keep factor of 10 on grid
eps_m = mns.mylogspace(1e-2, 1e-9, nD)
lL_m = lL * (1 + eps_m)
err_k = []
for lL_k in lL_m:
    err_k.append(s2.error2D((lL_k, lDt), *eargs))
    
dlL_m = lL_m - lL
derr_dlL = (np.array(err_k) - err0) / np.column_stack((dlL_m, dlL_m))
```

now same for time derivatives

```python
epsm = mns.mylogspace(1e-2, 1e-9, nD)
lDt_m = ldt * (1 + epsm)
err_k = []
for lDt_k in lDt_m:
    err_k.append(s2.error2D((lL, lDt_k), *eargs))
    
dlDt_m = lDt_m - ldt 
derr_dlDt = (np.array(err_k) - err0) / np.column_stack((dlDt_m, dlDt_m))
```

```python
f, ((ax1,ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True, figsize = (8,6))
deM_dlL = derr_dlL[:, 0]
ax1.loglog(epsm[deM_dlL > 0], deM_dlL[deM_dlL > 0], 'ob')
ax1.loglog(epsm[deM_dlL < 0], -deM_dlL[deM_dlL < 0], 'og')
ax1.set_ylabel('DerrM / DlogL')

deL_dlL = derr_dlL[:, 1]
ax3.loglog(epsm[deL_dlL > 0], deL_dlL[deL_dlL > 0], 'ob')
ax3.loglog(epsm[deL_dlL < 0], -deL_dlL[deL_dlL < 0], 'og')
ax3.set_ylabel('DerrL / DlogL')
ax3.set_xlabel('stepsize [DlogL / logL]')

deM_dlDt = derr_dlDt[:, 0]
ax2.loglog(epsm[deM_dlDt > 0], deM_dlDt[deM_dlDt > 0], 'ob')
ax2.loglog(epsm[deM_dlDt < 0], -deM_dlDt[deM_dlDt < 0], 'og')
ax2.set_ylabel('DerrM / DlogDt')

deL_dlDt = derr_dlDt[:, 1]
ax4.loglog(epsm[deL_dlDt > 0], deL_dlDt[deL_dlDt > 0], 'ob')
ax4.loglog(epsm[deL_dlDt < 0], -deL_dlDt[deL_dlDt < 0], 'og')
ax4.set_ylabel('DerrL / DlogDt')
ax4.set_xlabel('stepsize [DlogDt / logDt]')

plt.tight_layout()
```

#### Roundoff error? (NO)

Could roundoff error for small stepsizes be affecting the noise in the derivates?  NO!  Consider that lL and lDt are exponentiated into ther physical (unlogged) values for use in the structure equations.  So consider $x = \log(y)$  and a stepsize $\epsilon$  in $x$ to $x' = (1 + \epsilon) x$  , i.e. $\Delta x/x = \epsilon$  .  For the exponentiated $y$  the stepsize is $\Delta y / y = y^\epsilon - 1 \approx \epsilon \ln(y) = \epsilon x \ln(10)$  .  Thus the fractional errors in the physical quantities $y$ are BIGGER thatn the log quantities $x$ since $x > 1/\ln(10)$ by a large amount.  So it's almost certainly not a roundoff issue.


```python
def logerr(y, eps):
    return y**eps - 1
    
y1 = 10**lL
y2 = 10**lDt
loglog(eps_m, logerr(y1, eps_m) / eps_m)
loglog(eps_m, logerr(y2, eps_m) / eps_m)
#loglog(eps_m, eps_m)
```

it doesn't look like this should be an issue, but lets check
