```python
delad27 = 2./7
Y03 = 0.3
a10 = 10.0
a100 = 100.0

Mc = 5.0 * Me
rhoc = 3.2
rc = (3 * Mc / (4 * numpy.pi * rhoc))**(1./3)

prms10 = params(Mc, rc, a10, delad27, Y03, gamma = gammafn(delad27), R = Rfn(Y03), \
              Cv = Cvfn(Y03, delad27), Pd = Pdisk(a10, mstar, FSigma, FT), \
              Td = Tdisk(a10, FT), kappa = kdust)

prms100 = params(Mc, rc, a100, delad27, Y03, gamma = gammafn(delad27), R = Rfn(Y03), \
              Cv = Cvfn(Y03, delad27), Pd = Pdisk(a10, mstar, FSigma, FT), \
              Td = Tdisk(a10, FT), kappa = kdust)
```

```python
atm10 = atmload('Mc5.0.npz', prms = prms10)
atm100 = atmload('Mc5.0.npz', prms = prms100)
```

```python
model10 = atm10[0]
param10 = atm10[1]
prof10 = atm10[2]

model100 = atm100[0]
param100 = atm100[1]
prof100= atm100[2]
```

```python
crit10 = critical(param10, prof10, model10)
crit100 = critical(param100, prof100, model100)
```

```python
param10 = crit10[0]
prof10 = crit10[1]
dt10 = crit10[2]

param100 = crit100[0]
prof100 = crit100[1]
dt100 = crit100[2]

t10 = 0 * numpy.ndarray(shape = (len(dt10)), dtype = float)
for i in range(len(t10)):
    t10[i] = sum(dt10[:i])
    
t100 = 0 * numpy.ndarray(shape = (len(dt100)), dtype = float)
for i in range(len(t100)):
    t100[i] = sum(dt100[:i])
```

```python
L10 = param10.L
L100 = param100.L

MB10 = param10.MB
MB100 = param100.MB

Matm10 = MB10 * Me - Mc
Matm100 = MB100 * Me - Mc

Mcdot10 = L10 * rc / (G * Mc)
Mcdot100 = L100 * rc / (G * Mc)

dM10 = (MB10[1:] - MB10[:-1]) * Me
dM100 = (MB100[1:] - MB100[:-1]) * Me

tatm10 = Matm10[:-1] / (dM10 / dt10[:-1])
tatm100 = Matm100[:-1] / (dM100 / dt100[:-1])

ttot10 = MB10[:-1] * Me / (dM10 / dt10[:-1])
ttot100 = MB100[:-1] * Me / (dM100 / dt100[:-1])
```

```python
def Mdotacc(a, Mp, regime):
    if regime == 'slow':
        return Mp**(2./3) * (3 * 10**10)**(-1) * (a/10)**(-3)
    elif regime == 'interm':
        return Mp**(2./3) * (1.4 * 10**7)**(-1) * (a/10)**(-2)
    if regime == 'fast':
        return Mp**(2./3) * (3 * 10**5)**(-1) * (a/10)**(-3./2)
```

```python
def RHill(Mp, a):
    return a * cmperau * (Mp / Msun)**(1./3)
mu = 2.35 * mp
k0 = 0.1

def Rp(Mp):
    return 10**9 * Mp**(1./3)

def Mcrit(Mp, a, regime):
    return (1. / (64 * pi**2) * Mdotacc(a, Mp, regime) * (Me / (365 * 24 * 3600)) / (Rp(Mp) * RHill(Mp * Me, a)) \
        * a * cmperau * 0.1 / (sigma * G**3 * Msun**(1./3)) * (kb/mu)**4)**(3./5)
```
