```python
import numpy
from utils.constants import Me, Re, gammafn, Rfn, Cvfn, Pdisk, Tdisk, kdust, \
     params, mufn
from Analytical.analytic_comp import analytic_sols as ansols
from utils.userpath import userpath
from utils.parameters import FSigma, FT, mstar
import matplotlib.pyplot as plt
from RadSGPoly import plot_helper as ph
from RadSGPoly.profiles_poly import atmload
from RadSGPoly.cooling_poly import critical
import matplotlib.cm as cm
```

```python
a = 10.0
Mc = 5.0
delad27 = 2./7
Y03 = 0.3

prms = params(Me, Re, a, delad27, Y03, gamma = gammafn(delad27), R = Rfn(Y03), \
              Cv = Cvfn(Y03, delad27), Pd = Pdisk(a, mstar, FSigma, FT), \
              Td = Tdisk(a, FT), kappa = kdust)
```

```python
atm = atmload('Mc5.0.npz', prms = prms)
```

```python
model, param, prof = atm[0], atm[1], atm[2]
```

```python
crit = critical(param, prof, model)
```

```python
paramcrit = crit[0]
profcrit = crit[1]
dtcrit = crit[2]
```

```python
i1 = 2
i2 = 50

r1, r2, r3 = profcrit.r[i1] / Re, profcrit.r[i2] / Re, profcrit.r[-1] / Re
P1, P2, P3 = profcrit.P[i1], profcrit.P[i2], profcrit.P[-1]
T1, T2, T3 = profcrit.t[i1], profcrit.t[i2], profcrit.t[-1]
m1, m2, m3 = profcrit.m[i1] / Me, profcrit.m[i2] / Me, profcrit.m[-1] / Me

MB1 = paramcrit.MB[i1]
MB2 = paramcrit.MB[i2]
MB3 = paramcrit.MB[-1]

RB1 = paramcrit.RB[i1]
RB2 = paramcrit.RB[i2]
RB3 = paramcrit.RB[-1]

rcb1 = paramcrit.rcb[i1]
rcb2 = paramcrit.rcb[i2]
rcb3 = paramcrit.rcb[-1]

PB1, TB1, Pcb1, Tcb1, Mcb1 = paramcrit.PB[i1], paramcrit.TB[i1], paramcrit.Pcb[i1], paramcrit.Tcb[i1], paramcrit.Mcb[i1]
PB2, TB2, Pcb2, Tcb2, Mcb2 = paramcrit.PB[i2], paramcrit.TB[i2], paramcrit.Pcb[i2], paramcrit.Tcb[i2], paramcrit.Mcb[i2]
PB3, TB3, Pcb3, Tcb3, Mcb3 = paramcrit.PB[-1], paramcrit.TB[-1], paramcrit.Pcb[-1], paramcrit.Tcb[-1], paramcrit.Mcb[-1]

M = paramcrit.MB
```

```python
filename = 'PTm_profiles_v2.pdf'


f, (ax1, ax2, ax3) = plt.subplots(3, sharex=True, figsize = (13./2, 7.43/2))


ax1.loglog(r1, P1, color = 'purple', label = r'$M_{\mathrm{tot}}=$' + str(MB1)[:4] + r'$M_{\oplus}$')
ax1.loglog(r2, P2, color = 'blue', label = r'$M_{\mathrm{tot}}=$' + str(MB2)[:4] + r'$M_{\oplus}$')
ax1.loglog(r3, P3, color = 'red', label = r'$M_{\mathrm{tot}}=$' + str(MB3)[:4] + r'$M_{\oplus}$')
ax1.set_ylabel(r'P(dyne cm$^{-2}$)')
ax1.yaxis.set_ticks([10**(-2), 10, 10**4, 10**7, 10**10])
ax1.set_title(r'$\nabla_{\mathrm{ad}}=2/7$, $\mu=2.35$, $a=10\mathrm{AU}$, $M_c=5 M_{\oplus}$')
ax1.legend(prop = {'size':8})

ax2.loglog(r1, T1, color = 'purple')
ax2.loglog(r2, T2, color = 'blue')
ax2.loglog(r3, T3, color = 'red')
ax2.set_ylabel(r'T(K)')
ax2.yaxis.set_ticks([10, 10**2, 10**3, 10**4])
#ax[1].yaxis.set_ticks([10**(-2), 10, 10**4, 10**7, 10**10])
ax3.semilogx(r1, m1, color = 'purple')
ax3.semilogx(r2, m2, color = 'blue')
ax3.semilogx(r3, m3, color = 'red')
ax3.loglog(RB1, MB1, '.', color = 'purple', label = r'$R_{\mathrm{Bondi}}$')
ax3.loglog(RB2, MB2, '.', color = 'blue')
ax3.loglog(RB3, MB3, '.', color = 'red')
ax3.loglog(rcb1, Mcb1, '^', ms = 3, color = 'purple', label = r'$R_{\mathrm{RCB}}$')
ax3.loglog(rcb2, Mcb2, '^', ms = 3, color = 'blue')
ax3.loglog(rcb3, Mcb3, '^', ms = 3, color = 'red')
ax3.yaxis.set_ticks([4, 6, 8, 10])
ax3.set_ylabel(r'm($M_{\oplus}$)')
ax3.set_xlabel(r'$r(R_{\oplus})$')
ax3.legend(prop = {'size':7})

f.subplots_adjust(hspace=0)

plt.savefig(userpath + '/figs/ModelAtmospheres/RadSelfGravPoly/' + filename)

```
