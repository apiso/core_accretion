```python
a = 60.0
a10 = 10.0
a30 = 30.0

Mc = 5.0
delad27 = 2./7
Y03 = 0.3

prms = params(Me, Re, a, delad27, Y03, gamma = gammafn(delad27), R = Rfn(Y03), \
              Cv = Cvfn(Y03, delad27), Pd = Pdisk(a, mstar, FSigma, FT), \
              Td = Tdisk(a, FT), kappa = kdust)

prms10 = params(Me, Re, a10, delad27, Y03, gamma = gammafn(delad27), R = Rfn(Y03), \
              Cv = Cvfn(Y03, delad27), Pd = Pdisk(a10, mstar, FSigma, FT), \
              Td = Tdisk(a10, FT), kappa = kdust)

prms30 = params(Me, Re, a30, delad27, Y03, gamma = gammafn(delad27), R = Rfn(Y03), \
              Cv = Cvfn(Y03, delad27), Pd = Pdisk(a30, mstar, FSigma, FT), \
              Td = Tdisk(a30, FT), kappa = kdust)
```

```python
atm = atmload('Mc5.0.npz', prms = prms)
atm10 = atmload('Mc5.0.npz', prms = prms10)
atm30 = atmload('Mc5.0.npz', prms = prms30)
```

```python
model, param, prof = atm[0], atm[1], atm[2]
model10, param10, prof10 = atm10[0], atm10[1], atm10[2]
model30, param30, prof30 = atm30[0], atm30[1], atm30[2]
```

```python
crit = critical(param, prof, model)
crit10 = critical(param10, prof10, model10)
crit30 = critical(param30, prof30, model30)
```

```python
paramcrit = crit[0]
profcrit = crit[1]
dtcrit = crit[2]

paramcrit10 = crit10[0]
profcrit10 = crit10[1]
dtcrit10 = crit10[2]

paramcrit30 = crit30[0]
profcrit30 = crit30[1]
dtcrit30 = crit30[2]
```
