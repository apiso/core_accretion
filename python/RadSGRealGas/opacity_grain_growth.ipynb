```python
j = 0
filelistS = []
arrS = 0 * np.ndarray(shape = (6, 1001))

for filename in os.listdir('Semenov_Rosseland')[1:]:
    filelistS = np.append(filelistS, filename)
    TS = []
    f = open(userpath + '/dat/Semenov_Rosseland/' + filename, 'r')
    i = 0
    for line in f:
        tempt = line.split()[0]
        temp = line.split()[1]
        TS = np.append(TS, tempt)
        arrS[j, i] = temp
        i = i + 1
    f.close()
    
    j = j + 1
```

```python
filename = 'kappa_Semenov_BL.pdf'

rnb = np.linspace(0, 255, 6)
j = 0
kBL = 0 * np.ndarray(shape = (6, len(TS)))

for i in ([2, 1, 0, 5, 4, 3]):
    for k in range(len(TS)):
        kBL[i, k] = kdustall(TS[k], 10**(-14 + 2 * j))
    pS, = plt.loglog(TS, arrS[i], c = cm.rainbow(numpy.int(rnb[j])))
    j = j + 1
    
l1 = plt.legend((r'$\rho=10^{-14}$', r'$\rho=10^{-12}$', r'$\rho=10^{-10}$', r'$\rho=10^{-8}$', \
r'$\rho=10^{-6}$', r'$\rho=10^{-4}$'), loc = 3, prop = {'size':10}, frameon = False)
    
j = 0
    
for i in ([2, 1, 0, 5, 4, 3]):    
    pBL, = plt.loglog(TS, kBL[i], '--', c = cm.rainbow(numpy.int(rnb[j])))
    j = j + 1

plt.legend([pS, pBL], ['Semenov', r'Bell & Lin'], loc = 8, prop = {'size':9}, frameon = False)
gca().add_artist(l1)
plt.xlabel('T [K]')
plt.ylabel(r'$\kappa$ [cm$^2$ g$^{-1}$]')

plt.savefig(userpath + '/figs/ModelAtmospheres/RadSelfGravRealEOS/Opacity/' + filename)
    
```
