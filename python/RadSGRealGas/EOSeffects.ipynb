```python
Y = 0.3
delad27 = 2./7
delad29 = 0.29677

a = 10.0
Mc = 10.0 * Me
rhoc = 3.2
rc = (3 * Mc / (4 * numpy.pi * rhoc))**(1./3)



prms = paramsEOS(Mc, rc, Y, a, Pd = Pdisk(a, mstar, FSigma, FT), \
                 Td = Tdisk(a, FT), kappa = kdust)
prmspoly2 = params(Mc, rc, a, delad27, Y, gamma = gammafn(delad27), R = Rfn(Y), \
              Cv = Cvfn(Y, delad27), Pd = Pdisk(a, mstar, FSigma, FT), \
              Td = Tdisk(a, FT), kappa = kdust)
prmspoly = params(Mc, rc, a, delad29, Y, gamma = gammafn(delad29), R = Rfn(Y), \
              Cv = Cvfn(Y, delad29), Pd = Pdisk(a, mstar, FSigma, FT), \
              Td = Tdisk(a, FT), kappa = kdust)
```

```python
atm = atmload('Mc10.0.npz', prms)
atmpoly = atmloadpoly('Mc10.0.npz', prmspoly)
atmpoly2 = atmloadpoly('Mc10.0.npz', prmspoly2)
atmcheck = atmload('Mc10.0_checktop_500K.npz', prms)
atmcheckb = atmload('Mc10.0_checkbottom_500K.npz', prms)

model, param, prof = atm
modelpoly, parampoly, profpoly = atmpoly
modelpoly2, parampoly2, profpoly2 = atmpoly2
modelcheck, paramcheck, profcheck = atmcheck
modelcheckb, paramcheckb, profcheckb = atmcheckb
```

```python
crit = critical(param, prof, model)
critpoly = criticalpoly(parampoly, profpoly, modelpoly)
critpoly2 = criticalpoly(parampoly2, profpoly2, modelpoly2)
critcheck = critical(paramcheck, profcheck, modelcheck, checktop = 1)
critcheckb = critical(paramcheckb, profcheckb, modelcheckb, checktop = -1)
```

```python
param = crit[0]
prof = crit[1]
dt = crit[2]

t = 0 * numpy.ndarray(shape = (len(dt)), dtype = float)
for i in range(len(t)):
    t[i] = sum(dt[:i])
    
    
parampoly = critpoly[0]
profpoly = critpoly[1]
dtpoly = critpoly[2]

tpoly = 0 * numpy.ndarray(shape = (len(dtpoly)), dtype = float)
for i in range(len(tpoly)):
    tpoly[i] = sum(dtpoly[:i])
    
parampoly2 = critpoly2[0]
profpoly2 = critpoly2[1]
dtpoly2 = critpoly2[2]

tpoly2 = 0 * numpy.ndarray(shape = (len(dtpoly2)), dtype = float)
for i in range(len(tpoly2)):
    tpoly2[i] = sum(dtpoly2[:i])
    

paramcheck = critcheck[0]
profcheck = critcheck[1]
dtcheck = critcheck[2]

tcheck = 0 * numpy.ndarray(shape = (len(dtcheck)), dtype = float)
for i in range(len(tcheck)):
    tcheck[i] = sum(dtcheck[:i])
    
    
paramcheckb = critcheckb[0]
profcheckb = critcheckb[1]
dtcheckb = critcheckb[2]

tcheckb = 0 * numpy.ndarray(shape = (len(dtcheckb)), dtype = float)
for i in range(len(tcheckb)):
    tcheckb[i] = sum(dtcheckb[:i])
```

```python
f, (ax1, ax2) = plt.subplots(2, sharex = True)

ax1.semilogy(param.MB, param.L, label = r'Real EOS')
ax1.semilogy(parampoly.MB, parampoly.L, label = r'Poly $\nabla_{\rm{ad}}=0.29677$')
#ax1.semilogy(parampoly2.MB, parampoly2.L, label = r'Poly $\nabla_{\rm{ad}}=2/7$')
ax1.semilogy(paramcheck.MB, paramcheck.L, label = r'Real EOS for T>500 K')
ax1.semilogy(paramcheckb.MB, paramcheckb.L, label = r'Real EOS for T<500 K')
ax1.set_ylabel(r'L [erg s$^{-1}$]')
ax1.set_title(r'$Y=0.3$, $a=10 \rm{ AU}$, $M_{\rm c}=10 M_{\oplus}$')
ax1.legend(prop = {'size':7})

ax2.semilogy(param.MB, t / (365 * 24 * 3600))
ax2.semilogy(parampoly.MB, tpoly / (365 * 24 * 3600))
#ax2.semilogy(parampoly2.MB, tpoly2 / (365 * 24 * 3600))
ax2.semilogy(paramcheck.MB, tcheck / (365 * 24 * 3600))
ax2.semilogy(paramcheckb.MB, tcheckb / (365 * 24 * 3600))
ax2.set_xlabel(r'$M_{\rm{tot}} [M_{\oplus}]$')
ax2.set_ylabel(r'$t$ [years]')

plt.tight_layout()
#f.subplots_adjust(hspace=0)

plt.savefig(userpath + '/figs/ModelAtmospheres/RadSelfGravRealEOS/EOSeffects/Ltplot.pdf')
```
