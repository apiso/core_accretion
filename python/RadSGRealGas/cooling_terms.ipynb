```python
def ordercb(param):
    if type(param.rcb[0]) == numpy.float64:
        return param
    else:
        for i in range(len(param)):
            param.Mcb[i] = sorted(param.Mcb[i])
            param.rcb[i] = sorted(param.rcb[i])
            param.Pcb[i] = sorted(param.Pcb[i], reverse = True)
            param.Tcb[i] = sorted(param.Tcb[i], reverse = True)
            param.Ucb[i] = sorted(param.Ucb[i])
            param.Egcb[i] = sorted(param.Egcb[i], reverse = True)
            param.Etotcb[i] = param.Egcb[i] + param.Ucb[i]
        
        return param

                    
```

```python
def taurcb(param, prof, model):
    tau = []
    
    for i in range(len(param)):
        if type(param.rcb[0]) == numpy.float64:
            rhorcb = 10**interplog10rho(log10(param.Tcb[i]), log10(param.Pcb[i]))
            Hrcb = param.Pcb / rhorcb * (param.rcb * Re)**2 / (G * param.Mcb * Me)
            krcb = interp_opacity(param.Tcb[i])
        else:
            rhorcb = 10**interplog10rho(log10(param.Tcb[i][-1]), log10(param.Pcb[i][-1]))
            Hrcb = param.Pcb[i][-1] / rhorcb * (param.rcb[i][-1] * Re)**2 / (G * param.Mcb[i][-1] * Me)
            krcb = interp_opacity(param.Tcb[i][-1])            
        
        tau = append(tau, Hrcb * rhorcb * krcb)
    
    return tau
```
